# Lists

```
friends = ["Kevin", "Karen", "Jim"]
friends[1] = "Mike" #replace the element from the list to the specific one
print(friends[1:]) #prints the list from the second element
```

```
lucky_numbers = ["4", "8", "15", "16", "23", "42"]
friends = ["Kevin", "Karen", "Jim", "Oscar", "Oscar", "Toby"]
friends.append("Creed") #add the element to the end of the list
friends.insert(2, "Kelly") #add the element to a specific place on the list
friends.extend(lucky_numbers) #add another list to the first one
print(friends)
friends.remove("Jim") #remove the specific element from the list
print(friends)
friends.pop() #remove the last element from the list
print(friends)
print(friends.index("Creed")) #shows the index of the element on the list (usefull for knowking if an specific element is on the list)
print(friends.count("Oscar")) #count the number of the specific element on the list
friends.sort() #organize in alphabetical order
print(friends)
friends2 = friends.copy() #copy the elements from one list to another
print(friends2)
friends.clear() #remove all elements from the list
print(friends)
```

# If Statement

```
is_male = False
if is_male:
  print("You are a male")
else:
  print("You are not a male")
```

```
is_male = False
is_tall = True
if is_male and is_tall:
  print("You are a tall male")
elif is_male and not(is_tall):
  print("You are a short male")
elif not(is_male) and is_tall:
  print("You are not a male, but you are tall")
else:
  print("You are neither male nor tall")
```

```
def max_num(num1, num2):
  if num1 == num2:
    return 0
  if num1 > num2:
    return num1
  elif num1 < num2:
    return num2
result = max_num(2, 3)
print(result)
```

```
num1 = float(input("Enter the first number: "))
op = input("Enter the operator: ")
num2 = float(input("Enter the second number: "))

if op == "+":
  print(num1 + num2)
elif op == "-":
  print(num1 - num2)
elif op == "*":
  print(num1 * num2)
elif op == "/":
  print(num1 / num2)
else:
  print("Invalid operator")
```

# Exponent Function

```
def raise_to_power(base_num, pow_num):
  result = 1
  for index in range(pow_num):
    result = result * base_num
  return result
print(raise_to_power(2, 2))
```

# Translator

```
def translate(phrase):
  translation = ""
  for letter in phrase:
    if letter.lower() in "aeiou":
      if letter.isupper():
        translation = translation + "G"
      else:
        translation = translation + "g"
    else:
      translation = translation + letter
  return translation
print(translate(input("Enter a phrase: ")))
```

# Classes and Objects

```
class Student: #Class
  def __init__(self, name, major, gpa, is_on_probation):
    self.name = name
    self.majo = major
    self.gpa = gpa
    self.is_on_probation = is_on_probation
  def on_honor_roll(self):
    if self.gpa >= 3.5:
      return True
    else:
      return False

student1 = Student("Jim", "Business", 3.1, False) #Object
print(student1.gpa)
print(student1.on_honor_roll())
```
